# Distribution of Features

Numerical Float Features

```python
import matplotlib.pyplot as plt

numeric_cols = df_extended.select_dtypes(include=['float64']).columns
df_scaled = (df_extended[numeric_cols] - df_extended[numeric_cols].min()) / (df_extended[numeric_cols].max() - df_extended[numeric_cols].min())
axes = df_scaled.plot(kind='density', subplots=True, layout=(3, 4), figsize=(15, 10))
for ax, col in zip(axes.flatten(), numeric_cols):
    ax.set_title(col)
    ax.legend().set_visible(False)
plt.tight_layout()
plt.show()
```

Integar Features

```python
import seaborn as sns
numeric_cols = df_extended.select_dtypes(include=['int64']).columns.drop('profile_pics_count')
fig, axes = plt.subplots(3, 4, figsize=(15, 10))

for ax, col in zip(axes.flatten(), numeric_cols):
    sns.violinplot(y=df_extended[col], ax=ax)
    ax.set_title(col)

# Remove any unused subplots
for ax in axes.flatten()[len(numeric_cols):]:
    ax.remove()

plt.tight_layout()
plt.show()
```

```python
counts = df['profile_pics_count'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.bar(counts.index, counts.values)
plt.xlabel('Profile Picture Count')
plt.ylabel('Frequency')
plt.title('Distribution of Profile Picture Counts')
plt.show()
```

Categorical Features

```python
numeric_cols = df_extended.select_dtypes(include=['object']).columns.drop('interest_tags')
fig, axes = plt.subplots(3, 4, figsize=(15, 10))

for ax, col in zip(axes.flatten(), numeric_cols):
    counts = df_extended[col].value_counts()
    ax.bar(counts.index.astype(str), counts.values)
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=90)

# Remove any unused subplots
for ax in axes.flatten()[len(numeric_cols):]:
    ax.remove()

plt.tight_layout()
plt.show()
```
