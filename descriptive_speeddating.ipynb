# Preprocessing

Binning Spalten löschen

```python
# lösche alle spalten die mit d_ angfangen
df_processed = df.loc[:, ~df.columns.str.startswith('d_')]
# lösche mir die has_null spalte    
df_processed = df_processed.loc[:, ~df_processed.columns.str.startswith('has_null')]
```

```python
df_processed.info()
```

```python
# gib mir alle object spalten
object_columns = df_processed.select_dtypes(include=['object']).columns.tolist()
num_columns = df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()
df_processed[object_columns].head(10)
```

# Korrelation

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_numeric = df_processed[num_columns]
df_numeric = df_numeric.drop(columns=['match'], errors='ignore')
corr_matrix = df_numeric.corr()
 # ganze Koreelations anzeigen
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Korrelation Matrix')
plt.show()
```

# Distribution of Features

Numerical Float Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_viz = df_processed.filter(regex='pref|important').copy()
df_viz['gender'] = df_processed['gender']

pref_cols = df_viz.filter(regex='pref').select_dtypes(include=['int64', 'float64']).columns.tolist()
important_cols = df_viz.filter(regex='important').select_dtypes(include=['int64', 'float64']).columns.tolist()

n_rows = max(len(pref_cols), len(important_cols))
fig, axes = plt.subplots(n_rows, 2, figsize=(12, 4 * n_rows), squeeze=False)

# Definiere Farbzuweisung für Geschlechter (nur auf der linken Seite ändern)
gender_colors = {'male': '#1f77b4', 'female': '#ff7f0e'}

for i in range(n_rows):
    # Linke Spalte: 'pref' Density-Plot mit individuellen Farben für Männer und Frauen
    ax_left = axes[i, 0]
    if i < len(pref_cols):
        col = pref_cols[i]
        for g in df_viz['gender'].unique():
            subset = df_viz[df_viz['gender'] == g]
            color = gender_colors.get(g.lower(), None)
            sns.kdeplot(subset[col].dropna(), fill=True, label=g, ax=ax_left, color=color)
        ax_left.set_title(f'Density Plot für {col}')
        ax_left.set_xlabel(col)
        ax_left.set_ylabel('Dichte')
        ax_left.set_xlim(0, 100)
        ax_left.legend()
    else:
        ax_left.axis('off')

    # Rechte Spalte: 'important' Density-Plot (unverändert)
    ax_right = axes[i, 1]
    if i < len(important_cols):
        col = important_cols[i]
        for g in df_viz['gender'].unique():
            subset = df_viz[df_viz['gender'] == g]
            sns.kdeplot(subset[col].dropna(), fill=True, label=g, ax=ax_right)
        ax_right.set_title(f'Density Plot für {col}')
        ax_right.set_xlabel(col)
        ax_right.set_ylabel('Dichte')
        ax_right.set_xlim(0, 100)
        ax_right.legend()
    else:
        ax_right.axis('off')

plt.tight_layout()
plt.show()
```
