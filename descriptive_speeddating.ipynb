# Korrelation

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_numeric = df_extended[num_columns]
df_numeric = df_numeric.drop(columns=['match'], errors='ignore')
corr_matrix = df_numeric.corr()
 # ganze Koreelations anzeigen
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Korrelation Matrix')
plt.show()
```

```python
# show me top 10 collrelations of the matrix
top_corr = corr_matrix.unstack().sort_values(ascending=False).drop_duplicates()
top_corr = top_corr[top_corr < 1].head(10)
print("Top 10 Korrelationen:")
print(top_corr) 
```

### Importance / Prefernce Features seems to be the same

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_viz = df_processed.filter(regex='pref|important').copy()
df_viz['gender'] = df_processed['gender']

pref_cols = df_viz.filter(regex='pref').select_dtypes(include=['int64', 'float64']).columns.tolist()
important_cols = df_viz.filter(regex='important').select_dtypes(include=['int64', 'float64']).columns.tolist()

n_rows = max(len(pref_cols), len(important_cols))
fig, axes = plt.subplots(n_rows, 2, figsize=(12, 4 * n_rows), squeeze=False)

# Definiere Farbzuweisung fr Geschlechter (nur auf der linken Seite ndern)
gender_colors = {'male': '#1f77b4', 'female': '#ff7f0e'}

for i in range(n_rows):
    # Linke Spalte: 'pref' Density-Plot mit individuellen Farben fr Mnner und Frauen
    ax_left = axes[i, 0]
    if i < len(pref_cols):
        col = pref_cols[i]
        for g in df_viz['gender'].unique():
            subset = df_viz[df_viz['gender'] == g]
            color = gender_colors.get(g.lower(), None)
            sns.kdeplot(subset[col].dropna(), fill=True, label=g, ax=ax_left, color=color)
        ax_left.set_title(f'Density Plot fr {col}')
        ax_left.set_xlabel(col)
        ax_left.set_ylabel('Dichte')
        ax_left.set_xlim(0, 100)
        ax_left.legend()
    else:
        ax_left.axis('off')

    # Rechte Spalte: 'important' Density-Plot (unverndert)
    ax_right = axes[i, 1]
    if i < len(important_cols):
        col = important_cols[i]
        for g in df_viz['gender'].unique():
            subset = df_viz[df_viz['gender'] == g]
            sns.kdeplot(subset[col].dropna(), fill=True, label=g, ax=ax_right)
        ax_right.set_title(f'{col}')
        ax_right.set_xlabel(col)
        ax_right.set_ylabel('Dichte')
        ax_right.set_xlim(0, 100)
        ax_right.legend()
    else:
        ax_right.axis('off')

plt.tight_layout()
plt.show()
```

### Rating 

```python
attractive_partner_o = df_processed.filter(regex=r'^(attractive|sincere|intelligence|funny|ambition|shared_interests|sinsere|ambitous)(_partner|_o)?$').columns.tolist()

# Group columns into three lists based on their names
rated_by_partner = [col for col in attractive_partner_o if "_o" in col]
rating_partner = [col for col in attractive_partner_o if "_partner" in col]
self_assessment = [col for col in attractive_partner_o if col not in rated_by_partner + rating_partner]

# Determine the maximum number of features among the three groups
max_features = max(len(rated_by_partner), len(self_assessment), len(rating_partner))

fig, axes = plt.subplots(max_features, 3, figsize=(12, 4 * max_features))

for i in range(max_features):
    # Column 0: Rated by Partner
    if i < len(rated_by_partner):
        col = rated_by_partner[i]
        rel_counts = df_processed[col].dropna().astype(int).value_counts(normalize=True).sort_index()
        sns.barplot(x=rel_counts.index.astype(str), y=rel_counts.values, ax=axes[i, 0])
        axes[i, 0].set_xlabel(col)
        axes[i, 0].set_ylabel("Relative Frequency")
    else:
        axes[i, 0].axis('off')
    
    # Column 1: Selfassessment
    if i < len(self_assessment):
        col = self_assessment[i]
        rel_counts = df_processed[col].dropna().astype(int).value_counts(normalize=True).sort_index()
        sns.barplot(x=rel_counts.index.astype(str), y=rel_counts.values, ax=axes[i, 1])
        axes[i, 1].set_xlabel(col)
        axes[i, 1].set_ylabel("Relative Frequency")
    else:
        axes[i, 1].axis('off')
    
    # Column 2: Rating Partner
    if i < len(rating_partner):
        col = rating_partner[i]
        rel_counts = df_processed[col].dropna().astype(int).value_counts(normalize=True).sort_index()
        sns.barplot(x=rel_counts.index.astype(str), y=rel_counts.values, ax=axes[i, 2])
        axes[i, 2].set_xlabel(col)
        axes[i, 2].set_ylabel("Relative Frequency")
    else:
        axes[i, 2].axis('off')

# Set one header per column (only the first row gets the header)
axes[0, 0].set_title("Rated by Partner")
axes[0, 1].set_title("Selfassessment")
axes[0, 2].set_title("Rating Partner")

plt.tight_layout()
plt.show()
```

### Interests

```python
import numpy as np

interests = df_processed.filter(regex=r'^(sports|tv|tvsports|exercise|dining|museums|art|hiking|gaming|clubbing|reading|theater|movies|concerts|music|shopping|yoga)$').columns.tolist()
# Determine subplot grid
n = len(interests)
ncols = 4
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, nrows * 2.5))
axes = axes.flatten()

for i, col in enumerate(interests):
    # Calculate relative frequencies: count each value divided by total counts
    rel_counts = df_processed[col].value_counts(normalize=True).sort_index()
    axes[i].bar(rel_counts.index.astype(str), rel_counts.values)
    axes[i].set_title(col)
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Relative Frequency")

# Remove extra subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Expectation

```python
# Filter columns that start with "expected"
expected_cols = df_processed.filter(regex='^expected').columns.tolist()

# Number of expected columns
n = len(expected_cols)
fig, axes = plt.subplots(n, 1, figsize=(8, 4 * n))
if n == 1:
    axes = [axes]  # ensure axes is iterable

for ax, col in zip(axes, expected_cols):
    # Drop NAs und convert to int
    data = df_processed[col].dropna().astype(int)
    # Normierte Hufigkeiten und sortieren
    counts = data.value_counts(normalize=True).sort_index()
    
    # Dynamische x-Labels nur fr vorkommende Werte
    labels = counts.index.astype(str).tolist()
    
    # Barplot mit dynamischen Labels
    sns.barplot(x=labels, y=counts.values, ax=ax, order=labels)
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel("Relative Frequency")

plt.tight_layout()
plt.show()
```

### Outcome

```python
# We assume df_processed and x_labels are already defined.
# Convert x_labels from strings to ints for reindexing
x_labels_int = [int(x) for x in range(1, 11)]  # Assuming x_labels are from 1 to 10

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for ax, col in zip(axes, ['like', 'guess_prob_liked']):
    # Count occurrences as ints and reindex according to x_labels_int, filling missing values with 0.
    counts = df_processed[col].dropna().astype(int).value_counts().sort_index()
    counts = counts.reindex(x_labels_int, fill_value=0)
    
    ax.bar([str(v) for v in counts.index], counts.values)
    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
    
plt.tight_layout()
plt.show()
```
